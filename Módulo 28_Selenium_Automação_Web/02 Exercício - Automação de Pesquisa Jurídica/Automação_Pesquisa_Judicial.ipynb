```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time

import os
import pandas as pd

options = webdriver.ChromeOptions()
options.add_experimental_option("prefs", {
  "download.default_directory": r"C:\Users\thiag_vrxui9r\downloads",
  "download.prompt_for_download": False,
  "download.directory_upgrade": True,
  "safebrowsing.enabled": True
})

navegador = webdriver.Chrome(options=options)
caminho = os.getcwd()

base_processos = pd.read_excel('Processos.xlsx')


#abrindo a página da web
arquivo = caminho + '\index.html'
navegador.get(arquivo)

for i in range(len(base_processos)):
    navegador.refresh()
    
    #aplicando Action Chains à página
    actions = ActionChains(navegador)

    botao_escolha_o_estado = navegador.find_element(By.XPATH, '/html/body/div/div/button')

    distrito_federal = navegador.find_element(By.XPATH, '/html/body/div/div/div/a[1]')
    rio_de_janeiro = navegador.find_element(By.XPATH, '/html/body/div/div/div/a[2]')
    sao_paulo = navegador.find_element(By.XPATH, '/html/body/div/div/div/a[3]')

    actions.move_to_element(botao_escolha_o_estado).perform()

    cidade = base_processos['Cidade'][i]

    if cidade == 'Rio de Janeiro':
        rio_de_janeiro.click() 
    elif cidade == 'Distrito Federal':
        distrito_federal.click()
    else:
        sao_paulo.click()
        
    pagina_inicial = navegador.window_handles[0]
    pagina_pesquisa = navegador.window_handles[1]

    navegador.switch_to.window(pagina_pesquisa)

    navegador.find_element(By.XPATH, '//*[@id="nome"]').send_keys(base_processos['Nome'][i])
    navegador.find_element(By.XPATH, '//*[@id="advogado"]').send_keys(base_processos['Advogado'][i])
    navegador.find_element(By.XPATH, '//*[@id="numero"]').send_keys(base_processos['Processo'][i])

    navegador.find_element(By.XPATH, '//*[@id="formulario"]/div/button').click()
    
    alerta = navegador.switch_to.alert
    alerta.accept()
        
    time.sleep(5)
    alerta = navegador.switch_to.alert
    texto_alerta = alerta.text
    
    if texto_alerta == 'Processo encontrado com sucesso! Deseja acessar?':
        texto_alerta = 'Processo encontrado com sucesso!'
    else:
        pass
    
    base_processos.iloc[i, 4] = texto_alerta
    
    alerta.accept()
    
    navegador.close()
    navegador.switch_to.window(pagina_inicial)

navegador.quit()
base_processos.to_excel(r'Final_Base_Processos.xlsx')
print('Consulta Finalizada')
```
